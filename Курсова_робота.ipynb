## Теоретичні відомості

### Метод пошуку по симплексу

Метод пошуку по симплексу відноситься до категорії евристичних методів. Процес пошуку заснований на тому, що експериментальним зразком, який містить найменшу кількість точок, є регулярний симплекс.  Регулярний симплекс в N - мірному просторі - це многогранник, який містить N+1 рівновіддалених від нього вершини.  випадку 2-ох змінних - це правильний трикутник (рівносторонній). в трьохвимірному просторі - тетраедр.

В алгоритмі пошуку по симплексу використовується наступна властивість симплексу: Новий симплекс можна побудувати на будь-якій грані початкового симплексу шляхом переносу вибраної вершини на відповідну відстань вздовж прямої, проведеної через центр тяжіння інших вершин. Отримана таким чином точка стає вершиною нового симплексу, а стара (вибрана для побудови) вершина  симплекса виключається.

Таким чином, при переході до нового симплекса вимагається лише одне
обчислення значення цільової функції.
Початковим етапом симплексного пошуку є побудова регулярного симплекса в просторі незалежних змінних. Після побудови симплекса обчислюються значення цільової функції в кожній вершині симплекса. З вершини, якої відповідає максимальне значення цільової функції (в задачах мінімізації цільової функції), проводиться проектуюча пряма через центр тяжіння інших вершин.
Проводиться відображення обраної вершини щодо центра тяжіння вздовж даної прямої. Отримана точка стає вершиною нового відбитого симплекса, а стара (вибрана) вершина виключається.

f(x 2 ) = max{f(x1), f(x2), f(x3)},

Ітерації тривають до тих пір, поки:
1) Не буде покрита точка мінімуму;
2) Чи не почнеться циклічний рух по 2-ом чи більше симплексам.
	В таких випадках використовуються наступні правила:

Правило 1. Накриття точки мінімуму.
Якщо вершина, якій відповідає найбільше значення цільової функції, побудована на попередній ітерації, то замість неї береться вершина, якій відповідає наступне за величиною значення цільової функції.


Снимок1.PNG

Правило 2. Циклічний рух.
Якщо деяка вершина симплекса не виключається протягом більш, ніж M
ітерацій, необхідно зменшити розміри симплекса за допомогою коефіцієнта
редукції, і побудувати новий симплекс, вибравши в якості базисної точку, якій
відповідає мінімальне значення цільової функції.
Спендіш, Хекст і Хімсворт запропонували обчислювати M, як M = 1,65N + 0,05N^2, де N - розмірність завдання.

Правило 3. Критерій закінчення пошуку.
Пошук завершується, коли або 1) розміри симплекса стають досить малими або 2) різниці між значеннями цільової функції в вершинах симплекса менше заданої величини.


### Переваги та недоліки методу

Переваги:

1) Мала кількість вихідних даних: параметри величини симплекса t, параметри зменшення симплекса, параметри закінчення пошуку.

2) Невеликий вплив помилки обчислень значень цільової функції на ефективність пошуку, так як основними даними є максимальні значення цільової функції, а не мінімальні.

3) Простота розрахунків та логічної структури. невеликий обсяг програми обчислень на ЕОМ.

4) Невеликий обсяг використовуваної пам'яті ЕОМ.

Недоліки:

1) Необхідно введення масштабування, щоб значення змінних можна було порівняти за величиною, так як вершини симплекса залежать від одного масштабного множника.

2) Алгоритм досить повільно сходиться, так як в ньому не використовується інформація з попередньої ітерації.

3) Відсутність прискорення пошуку при проведенні пошуку на викривлених ділянках з вузьким і яром або хребтом. Якщо симплекс з якої-небудь причини зменшується, пошук триває вже тільки зі зменшеною величиною кроку.

Для побудови вихідного зразка формула побудови регулярного симплекса досить зручна, однак не існує особливої необхідності для збереження якості
регулярності в процесі пошуку.


### Степенева функція

$$ f(x_1, x_2) = (10(x_1 - x_2)^2 + (x_1 - 1)^2)^4) $$
$$(x1, x2) = (-1.2;0.0). N = 2, e = 0.5.$$

## Реалізація методу

### Імпорт бібліотек

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
```

### Оголошення констант

```python
x0 = [-1.2, 0]
N = 2
M = int(1.65*N + 0.05*N**2)
xc = []
e = 0.05
```

### Оголошення функцій

```python
def triangle():
   a = 2
   d0 = a * ((N + 1) ** (1 / 2) + N - 1) / (N * 2 ** (1 / 2))
   d1 = a * ((N + 1) ** (1 / 2) - 1) / (N * 2 ** (1 / 2))

   x1 = [x0[0] + d1, x0[1] + d0]
   x2 = [x0[0] + d0, x0[1] + d1]

   setx = []
   xvector = [x0, x1, x2]
   xmin = 0
   counter = 0
   while 1:
       if a < 2:
           d0 = a * ((N + 1) ** (1 / 2) + N - 1) / (N * 2 ** (1 / 2))
           d1 = a * ((N + 1) ** (1 / 2) - 1) / (N * 2 ** (1 / 2))
           x1 = [xmin[0] + d1, xmin[1] + d0]
           x2 = [xmin[0] + d0, xmin[1] + d1]
           xvector = [xmin, x1, x2]
       fxlist = [function(xvector[0]), function(xvector[1]), function(xvector[2])]
       for x in xvector:
           if function(x) == max(fxlist):
               xmax = x
           elif function(x) == min(fxlist):
               xmin = x
           else:
               xmead = x
       xnew = []
       xc = []
       for i in range(len(x0)):
           xc.append((xmin[i] + xmead[i])/N)
       for i in range(len(x0)):
           xnew.append(2*xc[i] - xmax[i])
       xvector = [xmin, xmead, xnew]

       if counter % M == 0:
           setx.append(xvector)

           for x in setx:
               for y in x:
                   for z in y:
                       unique = np.unique(setx, return_counts = True)
           for s in unique[1]:
               if s > 1:
                   a = a*0.5
                   print(a)
                   break
           print('minimum x: ', xmin)
           print('function: ', function(xmin))
           print(counter)
       print(function(xmax))
       if function(xmin) < e:
           print('minimum x: ', xmin)
           print('function: ', function(xmin))
           print(counter)
           break
       counter = counter + 1

   return setx
```

### Виконання розрахунків та відображення результатів


```python
arrays = []
setx = triangle()
for i in range(len(setx)):
   arrays.append(np.array(setx[len(setx) - i - 1]))

plt.figure()
plt.scatter(arrays[len(arrays) - 1][:,0], arrays[len(arrays) - 1][:,1], s = 10, color = 'black')

for i in arrays:
   t1 = plt.Polygon(i[:3,:], color = 'black', fill=False)
   plt.gca().add_patch(t1)
plt.show()
```
